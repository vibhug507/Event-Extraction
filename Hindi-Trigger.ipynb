```python
import pickle
file = open('result.pkl','rb')
train_x = pickle.load(file)
train_y = pickle.load(file)
test_x = pickle.load(file)
test_y = pickle.load(file)
file.close()
```

```python
temp=train_x[:]+test_x[:]
te=[]
for i in temp:
    for j in i:
        te.append(j)
print(len(te))
```

```python
vocab=list(set(te))
print(len(vocab))
```

```python
print(vocab)
```

```python
word_index = {}
for i in range(len(vocab)):
    word_index[vocab[i]] = i + 1
```

```python
word_index
```

```python
train_X, test_X = train_x, test_x

for i in range(len(train_X)):
    for j in range(len(train_X[i])):
        train_X[i][j] = word_index[train_X[i][j]]
        
for i in range(len(test_X)):
    for j in range(len(test_X[i])):
        test_X[i][j] = word_index[test_X[i][j]]
```

```python
test_X[0]
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
max_len = 200

train_X = pad_sequences(train_X, maxlen = 200, padding = 'post')
test_X = pad_sequences(test_X, maxlen = 200, padding = 'post')
```

```python
train_Y, test_Y = train_y, test_y
```

```python
c1 = []
for tag in test_y:
    for i in tag:
        if i != None:
            c1.append(i)
print(len(set(c1)))
```

```python
c2 = []
for tag in train_y:
    for i in tag:
        if i != None:
            c2.append(i)
print(len(set(c2)))
```

```python
uniq_event = list(set(c1)) + list(set(c2))
```

```python
uniq_event_type = list(set(uniq_event))
```

```python
uniq_event_type.append(None)
```

```python
print(len(uniq_event_type))
print(uniq_event_type)
```

```python
event_index = {}
for i in range(len(uniq_event_type) - 1):
    event_index[uniq_event_type[i]] = i + 1
```

```python
for i in range(len(train_Y)):
    for j in range(len(train_Y[i])):
        if train_Y[i][j] is None:
            train_Y[i][j] = 0
        else:
            train_Y[i][j] = event_index[train_Y[i][j]]

for i in range(len(test_Y)):
    for j in range(len(test_Y[i])):
        if test_Y[i][j] is None:
            test_Y[i][j] = 0
        else:
            test_Y[i][j] = event_index[test_Y[i][j]]
```

```python
train_Y = pad_sequences(train_Y, maxlen = 200, padding = 'post')
test_Y = pad_sequences(test_Y, maxlen = 200, padding = 'post')
```

```python
train_X2 = []
test_X2 = []

for i in range(len(train_X)):
    l = []
    for j in range(len(train_X[i])):
        index = train_X[i][j]
        x = np.zeros(len(vocab))
        x[index - 1] = 1;
        l.append(x)
    train_X2.append(l)
        
for i in range(len(test_X)):
    l = []
    for j in range(len(test_X[i])):
        index = test_X[i][j]
        x = np.zeros(len(vocab))
        x[index - 1] = 1;
        l.append(x)
    test_X2.append(l)
```

```python
len(train_Y[0])
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import models
```

```python
len(train_X2)
```

```python
learning_rate = 0.01
vocab_size = len(vocab)

input_shape = (None, )

model = tf.keras.models.Sequential([
    layers.Embedding(input_dim=vocab_size, output_dim=300, input_shape=input_shape, trainable=True),
    layers.Conv1D(filters=200, kernel_size=2, activation='relu'),
    layers.MaxPooling1D(pool_size=4),
    layers.Dropout(rate=0.5),
    layers.Conv1D(filters=200, kernel_size=3, activation='relu'),
    layers.MaxPooling1D(pool_size=4),
    layers.Dropout(rate=0.5),
    layers.Conv1D(filters=200, kernel_size=4, activation='relu'),
    layers.MaxPooling1D(pool_size=4),
    layers.Dropout(rate=0.5),
    layers.LSTM(200),
    layers.Dense(200, activation='relu')
])

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

model.fit(train_X, train_Y, epochs=300, batch_size=32)
```

```python
print("Accuracy: 54%\n")
print("Hyperparamters: \n", "Epochs: 300\n", "Optimizer: adam\n", "Loss: Mean Squared Error\n", "Batch Size: 32\n")
```
